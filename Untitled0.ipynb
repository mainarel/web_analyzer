# Подготовка данных
Для того, чтобы подготовить данные для входа в нашу нейросеть нужно провести их векторизацию. Для этого мы делаем длину каждого отзыва 10000 и заполняем пропуски нулями. Далее идет разбиение наборов на обучающий и тестировочный.

````python
def vectorize(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1
    return results

data = vectorize(data)
targets = np.array(targets).astype("float32")

test_x = data[:10000]
test_y = targets[:10000]

train_x = data[10000:]
train_y = targets[10000:]

````

````python
model = models.Sequential()

model.add(layers.Dense(50, activation = "relu", input_shape=(10000,)))
model.add(layers.Dropout(0.1, noise_shape=None, seed=None))
model.add(layers.Dense(50, activation = "relu"))
model.add(layers.Dropout(0.2, noise_shape=None, seed=None))
model.add(layers.Dense(50, activation = "relu"))


model.add(layers.Dense(1, activation = "sigmoid"))
model.summary()
# compiling the model
model.compile(
 optimizer = "adam",
 loss = "binary_crossentropy",
 metrics = ["accuracy"]
)

````

````python
results = model.fit(
 train_x, train_y,
 epochs= 2,
 batch_size = 500,
 validation_data = (test_x, test_y)
)
````

````python
# Making predictions
print (test_x)
predictions = model.predict(data[:1])

print (len(data[:1]))
print (predictions)
#print (data[0])
````
